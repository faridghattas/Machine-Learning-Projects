```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('/kaggle/input/sonar-dataset/sonar_dataset.csv', header=None)
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df[60].value_counts()
```

```python
df.groupby(60).mean()
```

```python
x = df.drop(columns=60, axis=1)
x
```

```python
y = df[60]
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, stratify=y, random_state=1)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
y_predict = model.predict(x_test)
y_predict
```

```python
y_test
```

```python
accu_training = model.score(x_train, y_train)
print('Accuracy on training data: ', accu_training)
```

```python
accu_test = model.score(x_test, y_test)
print('Accuracy on test data: ', accu_test)
```
